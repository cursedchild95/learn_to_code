```python
def embed(alpha, X, pn,mid_band):
    k = 0
    for i in range(0,X.shape[0]):
        for j in range(0,X.shape[1]):
            if(mid_band[i][j] == 1):
                X[i][j] = X[i][j] + alpha*pn[k]
    return X
    
    
```

```python
def extract(X, midband):
    extracted_list = []
    for i in range(0,X.shape[0]):
        for j in range(0,X.shape[1]):
            if midband[i][j]==1:
                extracted_list.append(X[i][j])
    return extracted_list 
```
